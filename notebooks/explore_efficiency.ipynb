```python
from orderbike.metrics import directness, global_efficiency
import utg.create_graph as cg
from matplotlib import pyplot as plt
```

```python
graph_list = {}
graph_list["grid"] = cg.create_grid_graph(rows=10, cols=10, width=100, diagonal=False)
graph_list["radio_concentric"] = cg.create_concentric_graph(
    radial=10, zones=9, center=True, radius=100
)
graph_list["three_bridges"] = cg.create_bridge_graph(
    outrows=3, sscols=4, bridges=3, blength=300, block_side=100
)
graph_list["grid_with_diagonal"] = cg.create_grid_graph(
    rows=10, cols=10, width=100, diagonal=True
)
ge = [global_efficiency(G) for G in graph_list.values()]
di = [directness(G) for G in graph_list.values()]
```

```python
for idx, graph_name in enumerate(graph_list.keys()):
    plt.scatter(ge[idx], di[idx], s=60, zorder=2, label=graph_name)
plt.plot([0, 1], [0, 1], zorder=1, color="lightgrey")
plt.xlim([0.7, 1])
plt.ylim([0.7, 1])
plt.xlabel("Global efficiency")
plt.ylabel("Directness")
plt.legend()
```

```python
for idx, graph_name in enumerate(graph_list.keys()):
    plt.scatter(di[idx], ge[idx] - di[idx], s=60, zorder=2, label=graph_name)
plt.ylabel("Difference between global efficiency and directness")
plt.xlabel("Directness")
plt.legend()
```
